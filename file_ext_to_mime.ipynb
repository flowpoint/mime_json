```python
import pandas as pd
url = r'https://www.sitepoint.com/mime-types-complete-list/'
tables = pd.read_html(url) # Returns list of all tables on page
```

```python
ext_dict = tables[0].set_index('Suffixes applicable').to_dict()
```

```python
ext_dict['Media type and subtype(s)']
```

```python
with open('ext_to_mime.json', 'w') as ext_json:
    json.dump(ext_dict, ext_json)
```
